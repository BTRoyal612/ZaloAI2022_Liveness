```python
import csv
import shutil
import os
```

```python
label_file = "D:/ZaloAI/LIVENESS/train/train/label.csv"
videos = r'D:\ZaloAI\LIVENESS\train\train\videos'
fake = r'D:\ZaloAI\LIVENESS\dataset\fake_vids'
real = r'D:\ZaloAI\LIVENESS\dataset\real_vids'
```

```python
label = open(label_file)
label = csv.reader(label)
next(label)
```

```python
for row in label:
    target = ""
    original = os.path.sep.join([videos, row[0]])
    if row[1] == '1':
        target = real
    else:
        target = fake
    shutil.move(original, target)
    print(original)
    print(target)
```
